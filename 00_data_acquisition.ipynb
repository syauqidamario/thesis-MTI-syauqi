```python
import yfinance as yf
import pandas as pd
```

```python
# --- AMBIL DATA BBCA (Yahoo Finance) ---
bbca = yf.download("BBCA.JK", start="2020-01-01", end="2025-12-31")
bbca.to_csv('bbca_historical_price_5years.csv')
```

```python
# --- AMBIL DATA KURS USD/IDR (Ekonomi Makro) ---
# Kurs adalah indikator makro yang paling dinamis dan tersedia di Yahoo Finance
kurs = yf.download("IDR=X", start="2020-01-01", end="2025-12-31")[['Close']]
kurs.columns = ['Kurs']
kurs.to_csv('ekonomi_makro_indonesia.csv')
```
